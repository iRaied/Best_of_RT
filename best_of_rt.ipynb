```python
# List of dictionaries to build file by file and later convert to a DataFrame
df_list = []
folder = 'rt_html'
for movie_html in os.listdir(folder):
    with open(os.path.join(folder, movie_html)) as file:
        soup = BeautifulSoup(file, 'lxml')
        title = soup.find('title').contents[0][:-len(' - Rotten Tomatoes')]
        div_container = soup.find('div', class_="meter-value")
        audience_score = div_container.find('span', class_="superPageFontColor").contents[0][:-len('%')]
        div_container = soup.find('div', class_="audience-info hidden-xs superPageFontColor")
        number_of_audience_ratings = div_container.find_all('div')[1].contents[2].strip()
        # Append to list of dictionaries
        df_list.append({'title': title,
                        'audience_score': int(audience_score),
                        'number_of_audience_ratings': number_of_audience_ratings})
df = pd.DataFrame(df_list, columns = ['title', 'audience_score', 'number_of_audience_ratings'])
```
